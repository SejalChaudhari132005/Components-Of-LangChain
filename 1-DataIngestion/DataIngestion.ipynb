```python
## Web based loader
from langchain_community.document_loaders import WebBaseLoader
import bs4
loader=WebBaseLoader(web_paths=("https://www.thetalentedindian.com/splash-of-colours-meet-the-phenomenal-watercolour-artists-of-india/",),
                bs_kwargs=dict(parse_only=bs4.SoupStrainer(
                            class_=("mvp-post-title left entry-title","mvp-author-info-wrap left relative","mvp-post-main-out left relative")
                         ))
     ) 
```

```python
loader.load()
```
