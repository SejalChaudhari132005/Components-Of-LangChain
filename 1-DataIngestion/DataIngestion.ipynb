```python
## Creating a valid PDF file
from reportlab.lib.pagesizes import letter
from reportlab.pdfgen import canvas
import os

# Create a sample PDF with content
pdf_path = r'c:\Users\Sejal Pc\Desktop\Langchain\1-DataIngestion\safetyLLM.pdf'

# Create PDF with reportlab
c = canvas.Canvas(pdf_path, pagesize=letter)
c.drawString(100, 750, "Safety Guidelines for Large Language Models")
c.drawString(100, 730, "")
c.drawString(100, 710, "1. Data Privacy and Security")
c.drawString(100, 690, "   - Ensure all user data is encrypted and stored securely")
c.drawString(100, 670, "   - Follow GDPR and data protection regulations")
c.drawString(100, 650, "")
c.drawString(100, 630, "2. Responsible AI Use")
c.drawString(100, 610, "   - Avoid generating harmful or biased content")
c.drawString(100, 590, "   - Implement content filtering and moderation")
c.drawString(100, 570, "")
c.drawString(100, 550, "3. Model Transparency")
c.drawString(100, 530, "   - Clearly indicate when AI is being used")
c.drawString(100, 510, "   - Provide explanations for AI decisions")
c.save()

print(f"PDF created successfully at: {pdf_path}")
print(f"File exists: {os.path.exists(pdf_path)}")
```
