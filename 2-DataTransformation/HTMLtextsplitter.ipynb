Split By HTML Header

```python
from langchain_text_splitters import HTMLHeaderTextSplitter

html_string = """
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Simple HTML Page</title>
</head>
<body>

<header>
    <h1>Welcome to My Website</h1>
</header>

<nav>
    <a href="#">Home</a> |
    <a href="#">About</a> |
    <a href="#">Contact</a>
</nav>

<main>
    <section>
        <h2>About This Page</h2>
        <p>This is a basic HTML-only template without any CSS.</p>
    </section>

    <section>
        <h2>Services</h2>
        <ul>
            <li>Web Development</li>
            <li>Design</li>
            <li>Consulting</li>
        </ul>
    </section>
</main>

<footer>
    <p>&copy; 2026 My Website</p>
</footer>

</body>
</html>
"""

header_to_split_on=[
    ("h1","Header 1"),
    ("h2","Header 2"),
    ("h3","Header 3")
]

html_splitter = HTMLHeaderTextSplitter(headers_to_split_on=header_to_split_on)
html_header_split=html_splitter.split_text(html_string)
html_header_split
```

```python
url = "https://susanchiang.com/blog/how-to-paint-a-rainy-day-scene-in-watercolor/"
headers_to_split_on = [
    ("h1","Header 1"),
    ("h2","Header 2"),
    ("h3","Header 3")
]

html_splitter = HTMLHeaderTextSplitter(headers_to_split_on)
header_splitter = html_splitter.split_text_from_url(url)
header_splitter
```
